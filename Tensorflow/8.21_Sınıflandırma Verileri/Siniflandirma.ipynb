### Sınıflandırma modeli

```python
# x ve y  değerlerini belirledik  .valuese ile diziye çevirdik 
y=dataFrame["Type"].values  #type veirsi y olacak
x=dataFrame.drop("Type", axis=1).values 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping  ## overfitting problemini önlemek için kullanılacak
```

model oluşturma 

```python
x_train.shape #kolon sayısı 30 ilk noronu 30 yapabiliriz ilk adımda 
```

#### kaç tane kolon varsa ilk katmana stun sayısı kadar nöron eklemek ilk adım olabilir 
#### ilk layer ile son layer olan 1 arasındaki sayılardan seçeceğimiz nöron sayısını
#### ara nörnda kullanabiliriz 30 15 15 1

```python

model = Sequential()
model.add(Dense(units=30,activation="relu")) # unit boron için 
model.add(Dense(units=15,activation="relu")) 
model.add(Dense(units=15,activation="relu")) 
model.add(Dense(units=1,activation="sigmoid")) # çıkış layerea da activation verdik bu sefer sigmoid iyidir 
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam")
```

Train etme 

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1)
#epoch uçok büyük vermek işlemi uzatır ve overfitting e yol açar 
# epoch u çok verdik
```

```python
modelKaybi=pd.DataFrame(model.history.history) # loss kayıtları 
```

```python
modelKaybi.plot() #epoch u çok verdiğimiz için vall los aşşağı doğru loss ile beraber giderken birden ayrılmaya başlamış 
# model kendi verileri ile uydurma yaptı 
#val_los saçmaladığı zaman durmalı 
```

## EarlyStopping
### model saçmalamaması için epocs durdurulmalı 

```python

model = Sequential()
model.add(Dense(units=30,activation="relu")) # unit boron için 
model.add(Dense(units=15,activation="relu")) 
model.add(Dense(units=15,activation="relu")) 
model.add(Dense(units=1,activation="sigmoid")) # çıkış layerea da activation verdik bu sefer sigmoid iyidir 
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
earlystopping = EarlyStopping(monitor = "val_loss",mode="min",verbose=1,patience=25)  # epocs saçmalamaya başladığı zaman epocs u durdurur
```

```python
#epoch u700 olarak kaldı early  stopping durduracak 
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1, 
          callbacks=[earlystopping]) # early stopping i verdik  callbasks olduğu için çoğul dizi olarak verdik 
```

```python
modelKaybi=pd.DataFrame(model.history.history) # loss kayıtları 
modelKaybi.plot()
```

## DropOut

```python
#0.5 den sonra drop out bizi uarır genelde vermeyiz gibi
#her katmandan sonra koyalım dropout u 
model = Sequential()
model.add(Dense(units=30,activation="relu")) # unit boron için 

model.add(Dropout(0.6))

model.add(Dense(units=15,activation="relu")) 
model.add(Dropout(0.6))

model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1,activation="sigmoid")) # çıkış layerea da activation verdik bu sefer sigmoid iyidir 

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
#epoch u700 olarak kaldı early  stopping durduracak 
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1, 
          callbacks=[earlystopping]) # early stopping i verdik  callbasks olduğu için çoğul dizi olarak verdik 
```

```python
modelKaybi=pd.DataFrame(model.history.history) # loss kayıtları 
modelKaybi.plot()
```

```python
tahminlerimiz =  model.predict(x_test)
```

```python
tahminlerimiz
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
classification_report(y_test , tahminlerimiz) 
```

```python
print(confusion_matrix(y_test,tahminlerimiz) )
```
